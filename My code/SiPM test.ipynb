# Extracting segmented data of SiPM single photon measurements

```python
#Initialising packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt(r"C:\Users\kento\Documents\University of Manchester\University Work\Fourth Year\MPhys\MPhys---VUV-Photosensor-Testing-for-Neutrino-and-Dark-Matter-Detectors\My code\SiPM_run_1.csv", delimiter=',', skip_header=1)
```

```python
waveforms = np.zeros((640, 3, 1000))

for i in range(1000):
    waveforms[:,:,i] = data[i*640:(1+i)*640,:]
```

```python
for i in range(1000):
    plt.plot(waveforms[:,1,i])
    plt.ylabel("Voltage (Volts)")
    plt.xlabel("Time (Nanoseconds)")
    plt.ylim(-0.004, 0.002)

# plt.savefig(r"C:\Users\kento\Documents\University of Manchester\University Work\Fourth Year\MPhys\MPhys---VUV-Photosensor-Testing-for-Neutrino-and-Dark-Matter-Detectors\My code\plots")
```
